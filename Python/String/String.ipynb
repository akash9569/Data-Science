```python
# len() - Returns the length of a string
text = "Python Programming"
print("Length:", len(text))
```

```python
# title() - Converts the first character of each word to uppercase
print("python programming language".title())
```

```python
# capitalize() - Capitalizes the first letter of the string
print("hello python".capitalize())
```

```python
# strip() - Removes leading and trailing whitespaces
print("   Python   ".strip())
```

```python
# replace() - Replaces a substring with another substring
print("I like Java".replace("Java", "Python"))
```

```python
# split() - Splits a string into a list
text = "apple,banana,orange"
print(text.split(","))
```

```python
# join() - Joins elements of a list into a string
fruits = ["apple", "banana", "cherry"]
print(", ".join(fruits))
```

```python
# find() - Returns the index of the first occurrence of a substring
print("Python Programming".find("Program"))
```

```python
# count() - Counts the occurrences of a substring
print("banana".count("a"))
```

```python
# startswith() - Checks if string starts with a specific substring
print("Python".startswith("Py"))
```

```python
# isalpha() - Returns True if all characters are alphabetic
print("Hello".isalpha())
```

```python
# isnumeric() - Returns True if all characters are numeric
print("12345".isnumeric())
```
