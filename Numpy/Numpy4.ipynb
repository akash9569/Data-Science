```python
import numpy as np
import time
```

## Speed Test

```python
# Python list
size = 1_000_000_00
list1 = list(range(size))
list2 = list(range(size))
start = time.time()
result = [x + y for x, y in zip(list1, list2)]
end = time.time()
print(end-start)
```

```python
# Numpy list
size = 1_000_000_00
arr1 = np.array(size)
arr2 = np.array(size)
start = time.time()
result = arr1 + arr2
end = time.time()
print(end-start)
```

```python
# Compare the time taken to multiply two Python lists vs. two NumPy arrays.

start = time.time()
result = [x * y for x, y in zip(list1, list2)]
end = time.time()
print(end-start)
```

```python
# Compare the time taken to multiply two Python lists vs. two NumPy arrays.

start = time.time()
result = arr1 * arr2
end = time.time()
print(end-start)
```
