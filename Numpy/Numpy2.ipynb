## Numpy Indexing and Slicing of Matrix

```python
arr = np.arange(1,31).reshape(6,5)
arr
```

```python
# Indexing
# Use row, column format → mat[row][col] or mat[row, col]
arr[0]
```

```python
# Slicing
# Use : to get ranges of rows/columns → mat[row_start:row_end, col_start:col_end]
slice=arr[3:6,3:5]
slice
```

## Array Operation

### Arithmatic Operations

```python
# Arithmetic operations in NumPy work element by element and make calculations faster and easier.
a1=np.array([1,2,3,4])
a2=np.array([5,6,7,8])
# a1 + a2
# a1-a2
a1*a2
a1/a2
a1//a2
a1**a2
```

### Broadcasting

```python
# Broadcasting means automatically matching array shapes so that arithmetic operations can be done without writing loops.

l=[10,20,30,40,50]
arr=np.array(l)
arr+10
```

### Deep and Shallow Copy

```python
# Shallow:
    # Creates a new object, but does not copy the actual data.
    # Both the original and the copy share the same data in memory.
    # Changing one will affect the other.

```

```python
# Deep:
    # Creates a new object and copies all data separately.
    # Both arrays are independent — changing one won’t affect the other.

```

```python
# Matrix operations in NumPy include addition, subtraction, multiplication, transpose, inverse, and determinant, used for linear algebra and data computation.

a=np.array([[1,2],[3,4]])
b=np.array([[6,7],[8,9]])

# Multiplication of this matrix
a@b
# np.dot(a,b)
```

```python
# Transpose Matrix
a.T
```

### Advance Array Operations

#### Stacking

```python
a=np.array([1,2,3,4])
b=np.array([5,6,7,8])
np.vstack((a,b))
```

```python
np.hstack((a,b))
```

```python
np.column_stack((a,b))
```

#### Splitting Array
