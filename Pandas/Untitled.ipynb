```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Example data
df = pd.DataFrame({
    "Age": [21,22,23,24,25],
    "Marks": [80,82,85,90,87]
})

# Line plot
plt.plot(df["Age"], df["Marks"])
plt.title("Age vs Marks")
plt.xlabel("Age")
plt.ylabel("Marks")
plt.show()

# Seaborn scatter
sns.scatterplot(x="Age", y="Marks", data=df)
plt.show()
```
