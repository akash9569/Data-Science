## Main Data Structure
- Series: 1D Labelled Array.
- DataFrame: 2D Labelled Array(Like .

## Series

```python
# Series Example
s = pd.Series([10, 20, 30, 40])
s
```

## Series Creation
- Using List
- Using Dictionary


```python
# Using List
Name = ["Akash", "Saloni", "Vikash", "Shweta", "Adii", "Shweta"]
Age = [22, 21, 22, 21, 22, 21]
pd.Series(Name)   # Simple Print the Data
pd.Series(Age, index=Name)     # Custom Index Using Name
pd.Series(Age, index=Name, name='Akash Ka Data')    # Give the Custom Name
```

```python
# Using Dictionary
Marks = {
    "Math": 70,
    "English": 80,
    "Science": 90,
    "Hindi": 100
}
Marks=pd.Series(Marks,name='Akash Ke Marks')
Marks
```

## Series Attribute
- Size: Return the Size of Series.
- dtype: Return the DataType of this Series.
- Name: Give the Custome name then Return the Name of this Series, Name Attribute also Return the Name(Name).
- is_unique: Find the Unique value of this Series.
- index: Return the Column Heading Name.
- values: Return the Presented Data in Series.

```python
# Size
Marks.size

# Dtype
Marks.dtype

# Name
Marks.name

# is_unique
Marks.is_unique

# Index
Marks.index

# Values
Marks.values
```

## Series Methods

```python
# Head & Tail
sd=pd.read_csv('marks.csv')
sd.head(5)
sd.tail(5)

# Sample: Returns the randomly Data.
sd.sample

# Value_counts
sd.value_counts

# Sort_values
sd.sort_index

# Sort_index
sd.sort_index
```

## Series Maths Method

```python
# sum(): Series ke saare values ka total return karta hai.
sd.sum()

# mean(): Series ka average (mean) return karta hai
sd.mean(numeric_only=True)

# median(): Series ka middle value (sorted data ka beech wala number) return karta hai
sd.median(numeric_only=True)

# mode(): Series me sabse zyada baar aane wali value(s) return karta hai
sd.mode()

# min(): Series ka smallest (minimum) value return karta hai
sd.min()

# max(): Series ka largest (maximum) value return karta hai
sd.max()

# std(): Values ki spread/dispersion measure (standard deviation) return karta hai
sd.std(numeric_only=True)

# var(): Series ke values ki variance (spread measure) return karta hai
sd.var(numeric_only=True)

# count(): Series me total valid (non-missing) values ki count return karta hai
sd.count()

# describe(): Series ki statistical summary return karta hai (count, mean, std, min, quartiles, max)
sd.describe()
```
